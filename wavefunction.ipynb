```python
import Majorana_main  as Maj
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
```

```python
def convert_to_parser(string):
    return string.split(' ')+['-wavefunction']
```

```python
oneline='-L=1 -mu=1 -lead_num=1 -lead_pos=LR -x_num=1024 -y_num=3 -energy -barrier_E=0 -dissipation=0'
# oneline='-L=1 -mu=1 -lead_num=1 -lead_pos=LR -SE -x_max=1.2 -dissipation=0.00 -barrier_E=0.0 -muVar=1 -muVar_fn=randlist_100.dat'
# oneline='-L=0.4 -mu=5 -lead_num=2 -lead_pos=LR -SE -coupling_SC_SM=1.5 -Vzc=10 -x_max=8 -y_min=-3 -y_max=3 -dissipation=0. -barrier_E=0.0 -Delta0=3'
```

```python
parser=Maj.argparse.ArgumentParser()
args=Maj.parse_arguments(parser,convert_to_parser(oneline))
```

```python
nw=Maj.Nanowire(args)
```

```python
trial=[0.9589,0.15]
fig,axs=plt.subplots(1,2,figsize=(6.8*2,4))
results=[]
for y,ax in zip(trial,axs):
    result=nw.wavefunction(0.6,y)
    Maj.plot_wavefunction(result,args,fig,ax)
    results.append(result)
```
